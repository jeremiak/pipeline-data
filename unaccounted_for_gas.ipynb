# what does unaccounted for gas mean?

What does the value in `PERCENT_UNACC_GAS` mean?

```python
import pandas as pd
```

```python
years = ['./annual_gas_distribution_2010_present/annual_gas_distribution_%s.xlsx' % y for y in range(2010, 2017)]
```

```python
all = pd.concat([pd.read_excel(f) for f in years])
```

```python
cols = ['REPORT_YEAR',
 'REPORT_NUMBER',
 'SUPPLEMENTAL_NUMBER',
 'OPERATOR_NAME',
 'OPERATOR_ID','OPERATOR_TYPE', 'KNOWN_LEAKS', 'PERCENT_UNACC_GAS']
```

in 2016, the "TRI - COUNTY NATURAL GAS" (`OPERATOR_ID` == 19641) agency reported an unaccounted gas percentage at 65.90%. that number seems really high, why is it like that?

```python
tri = all[all['OPERATOR_ID'] == 19641]
tri[['REPORT_YEAR','REPORT_NUMBER','OPERATOR_NAME','OPERATOR_ID','OPERATOR_TYPE', 'KNOWN_LEAKS', 'PERCENT_UNACC_GAS']]
```

looks like they don't normally have much unaccounted for gas, but there were 19 reported leaks. maybe that contributed to a lot of unnacounted for gas.

in 2016, the "NEW SUMMERFIELD GAS CO" (`OPERATOR_ID` == 13390) agency reported an unaccounted gas percentage at 57.38%, the second highest that year. what is that about?


```python
summerfield = all[all['OPERATOR_ID'] == 13390]
summerfield[['REPORT_YEAR','REPORT_NUMBER','OPERATOR_NAME','OPERATOR_ID','OPERATOR_TYPE', 'KNOWN_LEAKS', 'PERCENT_UNACC_GAS']]
```

this is interesting, there doesn't seem to be as strong of a relationship between unaccounted for gas and known leaks. what happened in 2015 and 2016 where there were no leaks but a yearly doubling of unaccounted for gas?
